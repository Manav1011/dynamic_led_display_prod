```python
import matplotlib.pyplot as plt
import base64
import io

data = {'AvgeSpeed': 0.3, 'AvgeTemp': -0.0, 'AvgeHum': -0.16, 'AvgeSr': -0.07}

plt.figure(figsize=(8, 4))
plt.plot(data.keys(), data.values(), marker='o', linestyle='-', color='b', linewidth=2, markersize=8, markerfacecolor='r', markeredgecolor='g', markeredgewidth=1)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
buffer = io.BytesIO()
plt.savefig(buffer, format='png')
plt.close()
image_base64 = base64.b64encode(buffer.getvalue()).decode('utf-8')
print(image_base64)
```
